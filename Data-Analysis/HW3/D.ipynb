```
item = sqlContext.sql("SELECT * FROM d5")
condition = sqlContext.sql("SELECT * FROM d_json")
```

```
from pyspark.sql.types import DoubleType
item = item.withColumn("price", item["price"].cast("double"))
item = item.withColumn("shipping", item["shipping"].cast("double"))
item = item.withColumn("item_condition_id", item["item_condition_id"].cast("double"))
item = item.withColumn("train_id", item["train_id"].cast("double"))

condition = condition.withColumn("price", condition["price"].cast("double"))
condition = condition.withColumn("shipping", condition["shipping"].cast("double"))
condition = condition.withColumn("item_condition_id", condition["item_condition_id"].cast("double"))
condition = condition.withColumn("train_id", condition["train_id"].cast("double"))

item.printSchema()
condition.printSchema()
```

```
display(item)
```

```
# Import graphframes (from Spark-Packages)

from graphframes import *

# Create Vertices (brand) and Edges (item)

itemVertices = item.withColumnRenamed("brand_name", "id").distinct()

brandEdges = condition.withColumnRenamed("name", "src").withColumnRenamed("price", "dst")
# This GraphFrame builds upon the vertices and edges based on our trips (flights)

```

```
display(itemVertices)
```

```
display(brandEdges)
```

```
g = GraphFrame(itemVertices, brandEdges)
```

```
brandEdges.cache()
itemVertices.cache()
```

```
print( "brand: %d" % g.vertices.count())
print( "item: %d" % g.edges.count())
```

```
display(g.edges.filter("src='Nike' and price>20").groupBy("src","dst").avg("item_condition_id").sort("avg(item_condition_id)"))
```

```
display(g.edges.filter("src='Nike' and price>10"))
```

```
display(g.edges.filter("src='Sephora' and price>10").groupBy("src","dst").avg("item_condition_id").sort("avg(item_condition_id)"))
```
