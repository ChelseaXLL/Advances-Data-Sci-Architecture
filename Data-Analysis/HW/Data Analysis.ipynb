```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

from pandas.tools.plotting import scatter_matrix
import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb
```

```python
iris= pd.read_excel('HW1.xlsx')
iris.plot(kind='scatter',x='COLLISION_TYPE',y='PERSON_COUNT',c=['darkgrey'],s=150)
```

```python
#We can see that the number of person who get hurt decreases as collusion_type increases.
sb.regplot(x='COLLISION_TYPE',y='PERSON_COUNT',data=iris,scatter=True)
```

```python
iris= pd.read_excel('HW1.xlsx')
iris.plot(kind='scatter',x='UNBELTED_OCC_COUNT',y='PERSON_COUNT',c=['darkgrey'],s=150)
```

```python
#We can see that the number of person who was got hurt  increases as the number of unbelted person increases.
sb.regplot(x='UNBELTED_OCC_COUNT',y='PERSON_COUNT',data=iris,scatter=True)
```

```python
#The weather correlate with the number of person who get injured
sb.jointplot(iris['WEATHER'],iris['PERSON_COUNT'],kind='reg')
plt.show()
```

```python
#weather also can affect people who got hurt
sb.jointplot(iris['WEATHER'],iris['INJURY_COUNT'],kind='reg')
plt.show()
```

```python
#whether unbelted or not can significantly affect the people when it comes to who are likely get injured
sb.jointplot(iris['UNBELTED_OCC_COUNT'],iris['INJURY_COUNT'],kind='reg')
plt.show()
```

```python
#unbelted driving also can affect the people who are likely to die in a crash, but not as strong as it do to the people who get injured
sb.jointplot(iris['UNBELTED_OCC_COUNT'],iris['FATAL_COUNT'],kind='reg')
plt.show()
```
